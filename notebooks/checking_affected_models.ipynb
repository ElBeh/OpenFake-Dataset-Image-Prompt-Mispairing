```python
from datasets import load_dataset
from IPython.display import display, Markdown
from PIL import Image
import io

affected_models = ['flux.1-dev', 'sd-1.5-dreamshaper', 'sd-3.5', 'sdxl-realvis-v5']

ds = load_dataset("ComplexDataLab/OpenFake", streaming=True, split="train")

for model in affected_models:
    display(Markdown(f"# Checking model: {model}"))
    display(Markdown("---"))

    filtered = ds.filter(lambda x: x['model'] == model)

    for i, sample in enumerate(filtered.take(10)):
        display(Markdown(f"## {model} - Sample {i+1}"))
        display(Markdown(f"**Prompt:** _{sample['prompt']}_"))

        if isinstance(sample['image'], dict) and 'bytes' in sample['image']:
            img = Image.open(io.BytesIO(sample['image']['bytes']))
        else:
            img = sample['image']

        display(img)
        display(Markdown("---"))

        if i >= 9:
            break
```
